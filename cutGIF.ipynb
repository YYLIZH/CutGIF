```python
def make_gif(filepath,video,scale,output,quality):
    # video = 123.mp4
    # scale = 640:320
    # output = 456.gif
    os.chdir(filepath)
    if quality=='high':
        command = 'ffmpeg -i {video} -filter_complex "[0:v] scale={scale},split [a][b];[a] palettegen=stats_mode=single [p];[b][p] paletteuse=new=1" {output}'.format(video=video,scale=scale,output=output)
        subprocess.call(command,shell=True)
    elif quality =='low':
        command = 'ffmpeg -i {video} -filter_complex "[0:v] scale={scale},split [a][b];[a] palettegen [p];[b][p] paletteuse" {output}'.format(video=video,scale=scale,output=output)
        subprocess.call(command,shell=True)

#make_gif(filepath,'momoko_cut.mp4','640:-1','momoko.gif','high')
```

```python
filename= os.listdir(filepath)
for name in filename:
    if name[-3:len(name)]=='mp4':
        new_name=name[0:-3]+'gif'
        make_gif(filepath,name,'640:-1',new_name,'high')
```
